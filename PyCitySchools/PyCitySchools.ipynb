```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
schoolCount = school_data_complete["school_name"].nunique()
studCount = school_data_complete["Student ID"].nunique()
totalBudget = school_data_complete["budget"].unique() 
avgMath = school_data_complete["math_score"].mean()
avgReading = school_data_complete["reading_score"].mean()

mathCount = school_data_complete[school_data_complete.math_score >= 70].count(axis='columns')
mathCount = mathCount.count()
pctMath = (mathCount/studCount) * 100

readCount = school_data_complete[school_data_complete.reading_score >= 70].count(axis='columns')
readCount = readCount.count()
pctRead = (readCount/studCount) * 100


overall = (avgMath + avgReading)/2


districtSummary = pd.DataFrame({
    "Total Schools":[schoolCount],
    "Total Students":[studCount],
    "Total Budget":[totalBudget.sum()],
    "Avereage Math Score":[avgMath],
    "Average Reading Score":[avgReading],
    "% Passing Math" : [pctMath],
    "% Passing Reading" : [pctRead],
    "%Overall Passing":[overall]
    
})
districtSummary
# Add formatting / rounding
```
